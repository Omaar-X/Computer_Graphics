**Consider the input points (1, 1) and (5, 12). Draw the ine using Midpoint Line**
**Generation Algorithm.**
**Advantages of the Mid-Point Algorithm**
**Like Bresenham's algorithm this is also efficient since it uses only integer math.**
**This makes it fast. It produces lines that look smooth and accurate. And the**
**algorithm is easy to understand and implement.**

```python
import matplotlib.pyplot as plt

def midPoint(X1, Y1, X2, Y2):
    points = []
    dx = X2 - X1
    dy = Y2 - Y1

    d = dy - (dx / 2)
    x, y = X1, Y1
    points.append((x, y))

    while x < X2:
        x += 1
        if d < 0:
            d += dy
        else:
            d += (dy - dx)
            y += 1
        points.append((x, y))
    return points

# Given input points
X1, Y1 = 1, 1
X2, Y2 = 5, 12
algo_points = midPoint(X1, Y1, X2, Y2)
print("Midpoint Algorithm Chosen Pixels:")
for p in algo_points:
    print(p)
plt.figure(figsize=(8, 6))
ax = plt.gca()

plt.plot([X1, X2], [Y1, Y2], color='blue', linewidth=1, label='Actual Line')

for (x, y) in algo_points:
    rect = plt.Rectangle((x - 0.5, y - 0.5), 1, 1, facecolor='purple', alpha=0.6)
    ax.add_patch(rect)


plt.grid(True, which='both', color='black', linewidth=0.5)
plt.xticks(range(0, max(X1, X2) + 5))
plt.yticks(range(0, max(Y1, Y2) + 5))
ax.set_aspect('equal', adjustable='box')

# Labels
plt.title("Midpoint Line Generation Algorithm\n(From (1, 1) to (5, 12))")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
