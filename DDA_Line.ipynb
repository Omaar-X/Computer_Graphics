**Task 1: Take user input for starting and ending points and draw a** **line using DDA**

```python
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    # Handle vertical line
    if dx == 0:
        m = float('inf')
    else:
        m = dy / dx

    x_points = []
    y_points = []

    # Case 1: |slope| <= 1
    if abs(m) <= 1:
        if x1 > x2:
            x1, y1, x2, y2 = x2, y2, x1, y1
        x = x1
        y = y1
        while x <= x2:
            x_points.append(round(x))
            y_points.append(round(y))
            x += 1
            y += m
    else:
        if y1 > y2:
            x1, y1, x2, y2 = x2, y2, x1, y1
        x = x1
        y = y1
        while y <= y2:
            x_points.append(round(x))
            y_points.append(round(y))
            y += 1
            x += 1/m

    plt.plot(x_points, y_points, marker='o', color='blue')
    plt.title('DDA Line Drawing (User Input)')
    plt.grid(True)
    plt.show()

# User Input
while True:
    try:
        x1 = int(input("Enter x1: "))
        y1 = int(input("Enter y1: "))
        x2 = int(input("Enter x2: "))
        y2 = int(input("Enter y2: "))
        break # Exit loop if inputs are valid integers
    except ValueError:
        print("Invalid input. Please enter integers only.")


dda_line(x1, y1, x2, y2)
```

**Task 3: Add options to change the color or thickness of the line**

```python
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2, color='blue', linewidth=1):
    dx = x2 - x1
    dy = y2 - y1
    m = dy / dx if dx != 0 else float('inf')
    x_points, y_points = [], []

    if abs(m) <= 1:
        if x1 > x2: x1, y1, x2, y2 = x2, y2, x1, y1
        x, y = x1, y1
        while x <= x2:
            x_points.append(round(x))
            y_points.append(round(y))
            x += 1
            y += m
    else:
        if y1 > y2: x1, y1, x2, y2 = x2, y2, x1, y1
        x, y = x1, y1
        while y <= y2:
            x_points.append(round(x))
            y_points.append(round(y))
            y += 1
            x += 1/m

    plt.plot(x_points, y_points, marker='o', color=color, linewidth=linewidth)
    plt.title('Custom DDA Line')
    plt.grid(True)
    plt.show()

# Input from user
x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))
color = input("Enter line color: ") or 'blue'
thickness = float(input("Enter line thickness: ") or 1)

dda_line(x1, y1, x2, y2, color, thickness)
```

**Exercise 1: Modify the DDA program to draw a triangle by connecting three** **vertices**

```python
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2, x_points, y_points):
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x = x1
    y = y1
    for _ in range(steps + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc

def draw_triangle():
    x_points = []
    y_points = []
    dda_line(150, 150, 450, 150, x_points, y_points)
    dda_line(150, 150, 300, 300, x_points, y_points)
    dda_line(450, 150, 300, 300, x_points, y_points)
    plt.figure(figsize=(6, 6))
    plt.plot(x_points, y_points, marker='o', linestyle='', color='blue')
    plt.title('Triangle using DDA Algorithm with round()')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

draw_triangle()
```

**Exercise 3: Draw a grid and visualize line drawing over it**

```python
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    x_points = []
    y_points = []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x = x1
    y = y1
    for _ in range(steps + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc
    return x_points, y_points

fig, ax = plt.subplots(figsize=(6,6))

# Set limits and grid
ax.set_xlim(0, 500)
ax.set_ylim(0, 500)
ax.set_xticks(range(0, 501, 50))
ax.set_yticks(range(0, 501, 50))
ax.grid(True, linestyle='--', color='gray')
ax.set_aspect('equal')

# Draw line using DDA
x_vals, y_vals = dda_line(50, 50, 450, 350)
ax.plot(x_vals, y_vals, 'ro-', label='DDA Line')

ax.set_title("Grid with DDA Line Drawing")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()

plt.show()
```
