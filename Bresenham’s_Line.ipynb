**Exercise 1: Draw a line from (0, 0) to (8, 5).**

```python
import matplotlib.pyplot as plt

x1, y1 = 0, 0
x2, y2 = 8, 5

dx = x2 - x1
dy = y2 - y1
d = 2 * dy - dx
x, y = x1, y1

x_points, y_points = [x], [y]

print("x\ty\td")
print(f"{x}\t{y}\t{d}")

for i in range(dx):
    if d < 0:
        d = d + 2 * dy
    else:
        y += 1
        d = d + (2 * dy - 2 * dx)
    x += 1
    print(f"{x}\t{y}\t{d}")
    x_points.append(x)
    y_points.append(y)

# show graph
plt.plot(x_points, y_points, marker='o', color='blue')
plt.title("Exercise 1: Line from (0,0) to (8,5)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

**Implement the line drawing algorithm for lines with a negative slope.**

```python
import matplotlib.pyplot as plt

x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

dx = abs(x2 - x1)
dy = abs(y2 - y1)
sx = 1 if x2 > x1 else -1
sy = 1 if y2 > y1 else -1
d = 2 * dy - dx

x, y = x1, y1
x_points, y_points = [x], [y]

print("x\ty\td")
for i in range(dx + 1):
    print(f"{x}\t{y}\t{d}")
    x_points.append(x)
    y_points.append(y)
    if d >= 0:
        y += sy
        d -= 2 * dx
    d += 2 * dy
    x += sx

plt.plot(x_points, y_points, marker='o', color='red')
plt.title(f"Exercise 3: Negative Slope Line ({x1},{y1}) → ({x2},{y2})")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

**1. Modify the algorithm to draw lines in all 8 octants.**

```python
import matplotlib.pyplot as plt

def bresenham(x1, y1, x2, y2, color):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1
    d = 2 * dy - dx
    x, y = x1, y1

    x_points, y_points = [], []
    for i in range(dx + 1):
        x_points.append(x)
        y_points.append(y)
        if d >= 0:
            y += sy
            d -= 2 * dx
        d += 2 * dy
        x += sx
    plt.plot(x_points, y_points, marker='o', color=color)

# draw 4 lines in different octants
bresenham(0, 0, 8, 5, 'blue')    # Octant 1
bresenham(0, 0, 5, 8, 'green')   # Octant 2
bresenham(0, 0, -8, 5, 'red')    # Octant 4
bresenham(0, 0, 5, -8, 'purple') # Octant 7

plt.title("Home Task 1: Lines in 8 Octants (Example 4 shown)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

**2. Compare output with DDA algorithm.**

```python
import matplotlib.pyplot as plt

def bresenham(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1
    d = 2 * dy - dx
    x, y = x1, y1
    x_points, y_points = [], []
    for i in range(dx + 1):
        x_points.append(x)
        y_points.append(y)
        if d >= 0:
            y += sy
            d -= 2 * dx
        d += 2 * dy
        x += sx
    plt.plot(x_points, y_points, marker='o', color='blue', label='Bresenham')

def dda(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    steps = abs(dx) if abs(dx) > abs(dy) else abs(dy)
    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1
    x_points, y_points = [], []
    for i in range(int(steps) + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc
    plt.plot(x_points, y_points, marker='x', color='red', label='DDA')

bresenham(0, 0, 8, 5)
dda(0, 0, 8, 5)
plt.title("Home Task 2: Bresenham vs DDA")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

**3. Count the number of pixels drawn for a diagonal line.**

```python
def bres(x0,y0,x1,y1):
    dx=abs(x1-x0); dy=abs(y1-y0)
    sx=1 if x1>=x0 else -1; sy=1 if y1>=y0 else -1
    err=dx-dy; x,y=x0,y0; pts=[(x,y)]
    while x!=x1 or y!=y1:
        e2=2*err
        if e2>=-dy: err-=dy; x+=sx
        if e2<=dx: err+=dx; y+=sy
        pts.append((x,y))
    return pts

pts=bres(0,0,10,10)
print('diagonal pixels =',len(pts))
import matplotlib.pyplot as plt
xs,ys=zip(*pts)
plt.figure(); plt.plot(xs,ys,marker='o')
plt.axis('equal'); plt.grid(True); plt.show()
```

**4. Extend the code to draw multiple lines using loops.**

```python
import matplotlib.pyplot as plt

def bresenham_all_octants(x0, y0, x1, y1):
    points = []
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    sx = 1 if x1 >= x0 else -1
    sy = 1 if y1 >= y0 else -1
    err = dx - dy
    x, y = x0, y0
    points.append((x, y))
    while (x != x1 or y != y1):
        e2 = 2 * err
        if e2 >= -dy:
            err -= dy
            x += sx
        if e2 <= dx:
            err += dx
            y += sy
        points.append((x, y))
    return points


lines = [
    (0,0,10,2),
    (0,0,8,5),
    (0,0,5,8),
    (2,10,12,0),
    (0,10,10,0),
    (5,5,-1,9)
]

plt.figure(figsize=(6,6))
for x0,y0,x1,y1 in lines:
    pts = bresenham_all_octants(x0,y0,x1,y1)
    xs, ys = zip(*pts)
    plt.plot(xs, ys, marker='o')
plt.title('Multiple Bresenham Lines ')
plt.axis('equal')
plt.grid(True)
plt.show()
```
